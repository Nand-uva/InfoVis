```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_path = 'DataSets/Duitsland_Tabel.csv'
data = pd.read_csv(file_path, delimiter=';')
```

```python
vig_reason_counts = data['vig_reason'].value_counts()
```

```python
plt.figure(figsize=(10, 7))
plt.pie(vig_reason_counts, labels=vig_reason_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Migration Reasons')
plt.show()
```
